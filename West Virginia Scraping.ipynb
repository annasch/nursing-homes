```python
all_data = []
for soup in [soup1, soup2, soup3]:
    data = soup.find_all('div', class_='bodyCells')
    all_data.append(data[0])

for data in all_data:
    print(data.text)
#data = all_data[0]

#print(data.text)
```

```python
# print(all_data)
list_of_tables = []

for data in all_data:
    i = 0
    table = []
    cells = data.find_all(class_=["pivotTableCellWrap cell-interactive", "pivotTableCellWrap cell-interactive tablixAlignCenter"])
    for index, cell in enumerate(cells):
        if index%20==0:
            col = [cell.text]
        else:
            col.append(cell.text)
        if index%20==19 or index==len(cells)-1:
            table.append(col)
    list_of_tables.append(table)
    
for table in list_of_tables:
    print(table)
    print('\n')
    
#print(html_string)
#import pandas as pd
#dfs = pd.read_html(html_string)
#df = dfs[0]
#print(df.head())
```

```python
import pandas as pd

dfs = []
extra = []

for table in list_of_tables:
    df = pd.DataFrame(table)
    df1 = df.iloc[:8,:]
    df2 = df.iloc[8:16, :]
    if(len(df.index)):
        extra = df.iloc[16:,:].values.tolist()
    #print(len(df.index))
    #print(df1)
    #print(df2)
    
    df1_T = df1.T
    #print(df1_T)
    df2_T = df2.T
    #print(df2_T)
    
    columns = ['county', 'facility', 'status', 'residents_pos', 'residents_recov', 'staff_pos', 'staff_recov', 'residents_deaths']
    df1_T.columns = columns
    df2_T.columns = columns
    
    df_T = df1_T.append(df2_T, ignore_index=True)
    
    # print(df_T)
    dfs.append(df_T)

#print(extra)
# df = pd.DataFrame(table)

#df
```

```python
## Deal with extra data ##
flat_extra = [item for sublist in extra for item in sublist]
flat_extra = [item for item in flat_extra if item is not None]

# print(flat_extra)

extra_table = []

for index, cell in enumerate(flat_extra):
    if index%3==0:
        col = [cell]
    else:
        col.append(cell)
    if index%3==2:
        extra_table.append(col)
        
# print(extra_table)

extra_df = pd.DataFrame(extra_table)

extra_df_T = extra_df.T

extra_df_T.columns = columns

dfs.append(extra_df_T)
```

```python
final_df = pd.concat(dfs, ignore_index=True)

final_df
```
